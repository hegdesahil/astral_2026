# Functions in python

Just like in a math class, we can also define **functions** to capture some kind of operation that happens to a variable. For example, a quadratic function might be mathematically represented as $f(x) = x^2$. We can equivalently make this function in python:

```python
def f(x):
	return x**2
```

You'll note a few key things here. First, the function largely looks like the mathematical expression (we have the name, `f`, followed by a parenthesis, within which the arguments of the function are given, in this case `x`). With Python, though, we have to tell the computer that we're introducing a function by using the special word `def` (you can see it's special because it changed color). Then, we use the regular function definition, but instead of an equals sign, we use a colon to denote to the computer that we're now going to explain what's in the function. Then, we tell the computer we want it to `return` something to us, with that something being the operation we specified --- namely, raising $x$ to the power of 2, which is achieved with two asterisks.

We can then use the function as follows:

```python
print(f(3))
```

However, we can add even more complexity to our functions and can introduce multiple arguments, for example. 

```python
def g(x,y):
	return x+y

print(g(2,3))
```

# Data Types in Python

So far, we have been working entirely with numbers (which are called integers and floats by the computer). You can tell what data type a variable is at any time using the "type()" command:

```python
x = 5
y = 6.0
print(type(x))
print(type(y))
```

While, at the very bottom of things, your data will always be numbers like this, Python's power comes in when you start looking at its other data types, which are primarily set up to contain numbers in an organized way. Here are some of the basic data types in Python:

 * Integers
 * Floats
 * Lists and Arrays (collections of items) 
 * Strings (contained in quotes "like this")

 We won't really have the time to get into all of these different types of data, but for now we'll focus on just one that takes things a step beyond the simple numbers we've been working with so far. 

 **Lists** and **Arrays** are data types that allow you to store collections of other variables, rather than just one. As you'll see, this is really powerful because it allows us to efficiently carry out operations on all of those values, rather than each individual variable one at a time. 

```python
# I create a list using square brackets like this:
my_list = [1,2,3,4]

# and access the elements from the list like this:
print(my_list[1])
```

Here, I've made a list named `my_list`, and accessed and printed out one of the elements of that list by putting square brackets next to the name of the list and a number indicating the position, or **index**, of the value I want. Here you'll see another important point about Python --- when accessing these values in a list, Python starts counting at 0, rather than 1, so the second element of the list is represented by index 1, as we saw above.

We can make this list even more useful by converting it into an **Array**. To do this, we'll use the `numpy` package for python. Below, we'll demonstrate how to `import` a package like numpy, give it a nickname, and then use it to convert the list we made into an array. In general, when we use a function from a package, we need to write the package name (or nickname) followed by a dot, before then calling the function name.

```python
import numpy as np # import ____ says what to import and then as __ allows you to set the nickname
my_array = np.array(my_list)

print(my_array**3)
```

As you can see, by converting our list into an array we were then able to cube every number in the array basically instantaneously. Maybe you could've done it in a comparable time for just four numbers, but imagine if our array had hundreds of numbers! *Note: this sort of operation is only possible because we made our list into an array. If it was still a list, we would not have been able to do that cubing!*

Arrays work similarly to lists, but have some nuanced differences which won't be important for our purposes. This means you can access elements in the array the same way that we did for the list!

# Plotting in python

Python is also a really powerful tool to help us visualize our data. To see how this can be done, we'll use another useful package, `matplotlib`, which gives you a bunch of functions that help generate plots! For example, in the cell below, let's plot a sine function.

```python
# first, import the package
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x,y)
```

Woah! New function alert! You'll notice that I also just used `np.linspace()`, a function which, unlike arange (which has you pick a start and stop and advances by integer (or multiples) steps in between), lets you pick a start, a stop, and a number of subdivisions, and then will try to space them evenly. Read it as "give me an array from 0 to 10 with 100 evenly spaced points and save that as a variable named `x`." Read line two as "Give me an array that contains the sine of each value in the `x` array and save that as a variable named `y`." 

Then, I passed those as the two arguments of a function called `plot` from matplotlib --- so I referenced it as `plt.plot` and voila, we got a sinusoidal graph! Therefore, all we need is a collection of `x` and `y` values, and we can immediately visualize the data!

# Reading in data from a file

In astronomy, we'll often be loading up data from an observation that was taken with a telescope, like the ones here on the mountain! In order to visualize and analyze that data (i.e., do science), we need to know how to convert it into a format that the computer can understand and that we can then work with. In general, you'll often encounter complex file types designed for storing image data, but for our purposes, we'll start with the most basic: a text file.

You might have noticed that this python notebook is not the only file on your sidebar -- we've also put in two text files (*note: the `.csv` file type just refers to a specific type of text file, where the values are separated by commas, making it a Comma-Separated Values file!). In the cell below, we'll read in the file named `pleiades_CMD.csv` using the numpy `loadtxt` function.

```python
pleiades_color, pleiades_magnitude = np.loadtxt('pleiades_CMD.csv', delimiter = ',').T

print(pleiades_color, pleiades_magnitude)
```

Ok so here we gave the function the file name in quotes and then in the second argument specified that the columns of the data were separated by (if we want to be fancy, *delimited* by) commas. Finally the `.T` after the closing parenthesis basically flips the rows and columns of the resulting array, but you don't need to worry about that for now.

If you compare what we printed out with the contents of the text file (which you can inspect by double clicking on it), you can check that things were read in alright.

## Exercise 5: color magnitude diagram
You might see that we've read in two columns of data from this file and suggestively labeled them *color* and *magnitude*. In fact, this data is from the [Gaia space telescope](https://www.esa.int/Science_Exploration/Space_Science/Gaia), which surveyed the positions and properties of billions of stars in our galaxy over the past decade. What you've just read in are the colors and brightnesses (magnitudes) of the stars from the Pleaides cluster (which you might have seen in the sky before!).

In the cell below, make a graph of the color (on the x axis) vs the magnitude (on the y axis). You'll want to use the `scatter` function from matplotlib, which works the same way as the `plot` function we just used except it just plots individual points.

Then, you'll want to make sure that your x and y axes are oriented in the correct direction --- remember, smaller magnitudes are brighter stars, so we want more negative numbers to be higher up. To do this, you can use the matplotlib `xlim` and `ylim` functions to specify the range of x and y values you want to show. 

You should also use the `xlabel` and `ylabel` functions to add labels to your axes. (If you'd like, you can also play around with the color of your points by adding an argument to the `scatter` function of `color = `, change the shape of the points with `marker =`, and/or the marker size with `s = `!)

```python
plt.scatter(pleiades_color, pleiades_magnitude, s = 40, color = 'purple', alpha = 0.5)
plt.ylim(15, -3)
plt.xlabel('color')
plt.ylabel('magnitude')
```

If all went well, you should see a collection of points following a diagonal line from the upper left to the lower right of your graph. What you've just made is what's called a color-magnitude diagram and the line you're seeing is called the *main sequence*, and represents the region of colors and magnitudes within which stars live for most of their lives. As you might have heard, once stars exhaust the hydrogen fuel in their cores, they start to expand and cool down, moving them away from this main sequence (they'd move up and to the right). Because all of the stars in this diagram lie nicely along the line, it tells us that the Pleiades is a relatively young cluster, without many stars that have evolved off of the main sequence. Pretty cool, right!
